```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
#load the data and assign columns

results = pd.read_csv(r"f1app\data\results.csv",sep = ',',encoding='utf-8')
races = pd.read_csv(r"f1app\data\races.csv",sep = ';',encoding='utf-8')
drivers = pd.read_csv(r"f1app\data\drivers.csv",sep = ',',encoding='utf-8')
constructors = pd.read_csv(r"f1app\data\constructors.csv",sep = ',',encoding='utf-8')
#results.head()
#races.head()
drivers.head()
#constructors.head()
```

```python
#merge datasets
df = pd.merge(results, races[["raceId", "year", "name", "round"]], on="raceId", how="left")
df = pd.merge(df, drivers[["driverId", "driverRef", "nationality"]], on = "driverId", how="left")
df = pd.merge(df, constructors[["constructorId", "name", "nationality"]], on="constructorId", how="left")
#df = pd.merge(results, races, on="raceId", how="left")
#df = pd.merge(df, drivers, on = "driverId", how="left")
#df = pd.merge(df, constructors, on="constructorId", how="left")
print(df)
```

```python
df.drop(["number", "position", "positionText", "laps", "fastestLap", "statusId", "resultId", "raceId", "driverId", "constructorId"], axis=1, inplace = True)
print(df)
```

```python
#rename columns
df.rename(columns={"rank":"fastestLapRank", "name_x":"gpName", "nationality_x":"driverNationality", "name_y":"constructorName", "nationality_y":"constructorNationality", "driverRef":"driver"}, inplace = True)
```

```python
#rearrange the columns
df = df[["year", "gpName", "round", "driver", "constructorName", "grid", "positionOrder", "points", "time", "milliseconds", "fastestLapRank", "fastestLapTime", "fastestLapSpeed", "driverNationality", "constructorNationality"]]
print(df)
```

```python
#drop incomplete seasons
df = df[df["year"]!=2020]
df = df[df["year"]!=2021]
df = df[df["year"]!=2022]
```

```python
#sort values
df = df.sort_values(by=["year", "round", "positionOrder"], ascending = [False, True, True])
```

```python
#replace \N values
df.time.replace("\\N", np.nan, inplace = True)
df.milliseconds.replace("\\N", np.nan, inplace = True)
df.fastestLapRank.replace("\\N", np.nan, inplace = True)
df.fastestLapTime.replace("\\N", np.nan, inplace = True)
df.fastestLapSpeed.replace("\\N", np.nan, inplace = True)
```

```python
#change datatypes
df.fastestLapSpeed = df.fastestLapSpeed.astype(float)
df.fastestLapRank = df.fastestLapRank.astype(float)
df.milliseconds = df.milliseconds.astype(float)
```

```python
#reset index
df.reset_index(drop=True, inplace = True)
```

```python
sb.set_palette("Set3")
plt.rcParams["figure.figsize"]=10,6
```

```python
#Gp Winners
driver_winner = df.loc[df["positionOrder"]==1].groupby("driver")["positionOrder"].count().sort_values(ascending=False).to_frame().reset_index()

#Barplot of all winners
sb.barplot(data=driver_winner, y ="driver", x="positionOrder", color="green", alpha=0.8)
plt.title("GP Winners in F1")
plt.ylabel("Drivers")
plt.xlabel("Number of Wins")
plt.yticks([])
```

```python
#create new datafram of top 10 Gp Winners
top10Drivers = driver_winner.head(10)
print(top10Drivers)
```

```python
sb.barplot(data = top10Drivers, y = "driver", x="positionOrder", color="blue", alpha = 0.8, linewidth=.8, edgecolor="black")
plt.title("Top 10 Gp Winners")
plt.ylabel("Drivers")
plt.xlabel("Number of Wins")
plt.savefig("top10winners.png")
```

```python
#constructor winners
constructor_winner = df.loc[df["positionOrder"]==1].groupby("constructorName")["positionOrder"].count().sort_values(ascending = False).to_frame().reset_index()

#barplot
sb.barplot(data = constructor_winner,y = "constructorName", x = "positionOrder", color = "red", alpha = 0.8, linewidth = .8, edgecolor="black")
plt.title("Constructors with most GP wins")
plt.ylabel("Constructors")
plt.xlabel("Numeber of wins")
plt.yticks([])
```

```python
#top 10 constructors
top10constructors = constructor_winner.head(10)
print(top10constructors)
```

```python
sb.barplot(data = top10constructors, y = "constructorName", x="positionOrder", color="blue", alpha = 0.8, linewidth=.8, edgecolor="black")
plt.title("Top 10 Constructors with most wins")
plt.ylabel("Constructors")
plt.xlabel("Number of Wins")
plt.savefig("top10constructors.png")
```

```python
df_no_zero = df[df["grid"] != 0]

plt.figure(figsize = [12,7])
sb.regplot(data = df_no_zero, x = "grid", y="positionOrder", x_jitter = 0.3, y_jitter = 0.3, scatter_kws = {"alpha": 1/5})
plt.title("Starting vs finishing positions")
plt.ylabel("Finish place")
plt.xlabel("Starting position")
plt.savefig("startvsfinish.png")
```

```python
#new data frame for seasons 2004 and onwards
df_speed = df[df["year"]>=2004]
df_group_speed = df_speed.groupby(["gpName", "year"])["fastestLapSpeed"].mean().to_frame().reset_index()

#create the facetgrid
g = sb.FacetGrid(data = df_group_speed, col = "gpName", col_wrap = 5)
g.map(plt.scatter, "year", "fastestLapSpeed", alpha = 0.8, linewidth = 0.8, edgecolor = "black", s = 100)
g.set_titles("{col_name}")
g.set_xlabels("Year")
g.set_ylabels("Average fastest speed (km/h)")
plt.subplots_adjust(top=0.92)
g.fig.suptitle("Average speed amongst all teams during the fastest lap at individual GPs")
plt.savefig("avgspeed.png")
```
